```python
data = data.dropna(subset=["Experiment ID"]) 

missing_values = data.isnull().sum()
print(missing_values)
```

```python
# Cell size missing? - perform median imputation
data_copy = data.copy()  # Makes a deep copy
cell_size_median = data_copy["Cell_size"].median()
data_copy["Cell_size"] = data_copy["Cell_size"].fillna(cell_size_median)
print("Median cell size ", cell_size_median)
data_copy["Cell_size"] = data_copy["Cell_size"].fillna(cell_size_median)


print()
missing_values = data_copy.isnull().sum()
print(missing_values)
```

## Imputation

```python

# Display the number of missing values in "X_cell_position" before imputation
print("Missing values in 'X_cell_position' before imputation:", data['X_cell_position'].isnull().sum())

# Calculate the mean of "X_cell_position", excluding missing values
mean_value = data['X_cell_position'].mean()

# Fill missing values in "X_cell_position" with the calculated mean
data['X_cell_position'].fillna(mean_value, inplace=True)

# Display the number of missing values in "X_cell_position" after imputation
print("Missing values in 'X_cell_position' after imputation:", data['X_cell_position'].isnull().sum())


```

## fILTER Columns



Filter and display the columns "Experiment ID", "Time stamp", and "Apparatus used" for all entries where the experiment happened in October 2024

```python


# Convert "Time stamp" to datetime format if it's not already
data['Time stamp'] = pd.to_datetime(data['Time stamp'], errors='coerce')

# Filter for records where the experiment date is in October 2024
filtered_data = data[(data['Time stamp'].dt.month == 10) & (data['Time stamp'].dt.year == 2024)]

# Display the specific columns
result = filtered_data[['Experiment ID', 'Time stamp', 'Apparatus used']]
print(result)
```
